```python
# Write the utility functions you need in your analysis.

### Function for calculating C-index ###
# y: array containing true label values.
# yp: array containing the predicted label values.

# this helper is from EMLM exercise 2
def cindex(y, yp):
    n = 0
    h_num = 0
    for i in range(0, len(y)):
        t = y[i]
        p = yp[i]
        for j in range(i+1, len(y)):
            nt = y[j]
            np = yp[j]
            if (t != nt):
                n = n + 1
                if (p < np and t < nt) or (p > np and t > nt):
                    h_num += 1
                elif (p == np):
                    h_num += 0.5
    return h_num/n



### TODO
### function to create groups of the data

```

#### Implement and run cross-validation

```python
# Implement and run the requested cross-validation. Report and interpret its results.
```

```python
#### ALGO in pseudo

# for each t types do:
# implement leave-one-out with the requirements for each run
# fit knn
# predict
# save score
# calculate c-index score from lists of true values and prediction
# save to a dict

def ten_nearest_neighbours(X,Y,P,TYPE_OF_OBSERVATION,nn=10) -> dict:

### SETUP ###
  loo = LeaveOneOut() # https://scikit-learn.org/dev/modules/generated/sklearn.model_selection.LeaveOneOut.html
  knn = KNeighborsRegressor(n_neighbors=nn) # nn count from the letter
  CINDEX_SCORES = {} # dict for scores, key is the type of observation
###


  for tt in TYPE_OF_OBSERVATION: # loop through the elements "A","B","C","D"
    print(f"type of observation is {tt}")
    TRUE_VALS = [] # actual test output values
    PRED_VALS = [] # values predicted by the KNearestNeigbour

    for train_index, test_index in loo.split(X): # loop through all possible splits
      X_train, X_test = X[train_index], X[test_index] # generate arrays from the selection of indices determined by the splitter in leave-one-out, default lengths are 399, 1
      Y_train, Y_test = Y[train_index], Y[test_index] # generate arrays from the selection of indices determined by the splitter in leave-one-out, default lengths are 399, 1

      # ********
      # Type A: both are in-sample-observations
      # no limitations to train set
      # no need to make modifications, keep the original arrays set above and just run fit and pred
      # ********

      # ********
      # Type B == 1: first of the pair is not included
      # S_train = {s : s in S except p_1(s) not in O_1}
      # The first pair members of the test observations are not allowed to appear in the training set
      if tt == "B":
        # its a list of lists with with two elements in the nested
        p_1 = P[test_index][0][0]
        # use list comprehension to get the indices from the train set that do not contain the same p_1
        new_train_indices = [i for i,pair in enumerate(P[train_index]) if pair[0] != p_1]
        X_train, Y_train = X[new_train_indices], Y[new_train_indices]
      # ********

      # ********
      # Type C == 2: second of the pair is not included
      # S_train = {s : s in S except p_2(s) not in O_2}
      # The second pair members of the test observations are not allowed to appear in the training set
      elif tt == "C":
        # its a list of lists with with two elements in the nested
        p_2 = P[test_index][0][1]
        #print(f"name of the p_2 in the test set: {p_2}")
        #for mole in P[train_index]:
        #  print(mole[1])
        # use list comprehension to get the indices from the train set that do not contain the same p_2
        new_train_indices = [i for i,pair in enumerate(P[train_index]) if pair[1] != p_2]
        X_train, Y_train = X[new_train_indices], Y[new_train_indices]
      # ********

      # ********
      # Type D == 3: neither element of the pair values is included
      # S_train = {s : s in S except p_1(s) not in O_1 except p_2(s) not in O_2}
      # None of the pair members of the test observations are allowed to appear in the training set
      elif tt == "D":
        new_train_indices = []
        # its a list of lists with with two elements in the nested
        p_1 = P[test_index][0][0]
        p_2 = P[test_index][0][1]
        # use loop to get the indices from the train set that do not contain the same p_1 or p_2
        for i, pa in enumerate(P[train_index]):
          p_case_1 = pa[0]
          p_case_2 = pa[1]
          if p_1 == p_case_1 or p_2 == p_case_2:
            continue
          else:
            new_train_indices.append(i)
        X_train, Y_train = X[new_train_indices], Y[new_train_indices]
      # ********

      # train
      knn.fit(X_train,Y_train)

      # predict
      Y_pred = knn.predict(X_test)

      # append to lists to use for cindex calculations
      TRUE_VALS.append(Y_test[0])
      PRED_VALS.append(Y_pred[0])

    c_index_value = cindex(y=TRUE_VALS, yp=PRED_VALS) # calculate the c_index_value from the y and y_pred
    print(f"C-index value: {c_index_value}")
    CINDEX_SCORES[tt] = c_index_value # add to the dict
  return CINDEX_SCORES

```

```python
%%time
# selection of training observations

## premises

# S = set of in-sample obseravation
# S_test = set of test observations
# p_1 = mapping from observations to their first pair members
# p_2 = mapping from observations to their second pair members

## Object sets
# O_1 = set of first pair members in S_test = {o : (exists s in S_set where p_1(s) = o)}
# O_2 = set of second pair members in S_test = {o : (exists s in S_set where p_2(s) = o)}

X = inputs_standardized.values
Y = outputs.values
P = pairs.values

TYPE_OF_OBSERVATION = ["A", "B", "C", "D"]

cind_scores = ten_nearest_neighbours(X=X,
                                     Y=Y,
                                     P=P,
                                     TYPE_OF_OBSERVATION=TYPE_OF_OBSERVATION)

```
